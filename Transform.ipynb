```python
import pandas as pd

# Set pandas to show all rows of the dataframes to analyse them easily
pd.set_option('display.max_rows', None)
```

```python
#Read CSV file for country_profile
df_country_profile = pd.read_csv("country_profile_variables.csv")

#Cleaning the Data
df_country_profile = df_country_profile.replace(to_replace="Democratic People's Republic of Korea", value="Congo")
df_country_profile
```

```python
#Read CSV file for World Happiness
df_world_happiness = pd.read_csv("2019.csv")
df_world_happiness.head()
```

```python

# Rename the column headers
world_happiness = df_world_happiness.rename(columns={"Country or region": "country"})
#world_happiness

world_happiness = world_happiness.replace(to_replace="Congo (Kinshasa)", value="Congo")

world_happiness 
                                                          
                                                    

```

```python
# Join both data in a dataframe
df_country_world_happiness_merged = pd.merge(world_happiness, df_country_profile,on='country', how='inner')
```

```python
df_country_world_happiness_merged
```

```python
#Generating the Data Load SQL file for Postgres

file = open("InsertData.sql","w")

for index, row in df_country_world_happiness_merged.iterrows():
    file.write("insert into world_happiness(country, score) values ('"+row['country']+"'', '"+str(row['Score'])+"');"+"\n")


file.close()
```
