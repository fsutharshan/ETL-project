```python
import pandas as pd
import re
import country_converter as coco
cc = coco.CountryConverter()
# Set pandas to show all rows of the dataframes to analyse them easily
pd.set_option('display.max_rows', None)
```

```python
#Read CSV file for country_profile
df_country_profile = pd.read_csv("country_profile_variables.csv")
df_country_profile
#Cleaning the Data
```

```python
for x,y in df_country_profile.iterrows():
    standard_name = cc.convert(names = y['country'], to = 'name_short')
    if standard_name == 'not found':
        print(y['country'])
    df_country_profile.loc[x, 'country']=standard_name

    
```

```python
df_country_profile
```

```python
#Read CSV file for World Happiness
df_world_happiness = pd.read_csv("2019.csv")
df_world_happiness.head()
```

```python
for x,y in df_world_happiness.iterrows():
    standard_name = cc.convert(names = y['Country or region'], to = 'name_short')
    df_world_happiness.loc[x, 'Country or region']=standard_name
```

```python

# Rename the column headers
world_happiness = df_world_happiness.rename(columns={"Country or region": "country"})

world_happiness
                                                          
                                                    

```

```python
# Join both data in a dataframe
df_country_world_happiness_merged = pd.merge(world_happiness, df_country_profile,on='country', how='left')
```

```python
df_country_world_happiness_merged
```

```python
#identify entries without matches

for x,y in df_country_world_happiness_merged.iterrows():
    if pd.isna(y['Individuals using the Internet (per 100 inhabitants)']):
        print(y['country'])
```

```python
cols=df_country_world_happiness_merged.columns
collumns = ""
for x in cols:
    hold = x.replace(" ","_")
    hold = hold.replace("," , "")
    hold = hold.replace("(","")
    hold = hold.replace(")","")
    hold =hold.replace("%" , "pct")
    hold = hold.replace(":", "")
    hold = hold.replace("/","")
    hold = hold.replace(".","")
    hold = hold.replace("=","_")
    hold = hold.replace("$","_dol")
    hold = hold.replace("+","plus")
    hold= hold.replace("-","_")
    hold=hold.lower()
    if hold != 'net_official_development_assist_received_pct_of_gni':
        collumns = collumns +hold+" varchar(30), "
    else:
        collumns = collumns +hold
f = open("columnnames.txt", "a")
f.write(collumns)
f.close()
print(collumns)
```

```python
regions = df_country_world_happiness_merged['Region'].unique()
regions_dfs={}
for x in regions:
    if pd.isna(x):
        print('ignore nan')
    else:
        print(x)
        regions_dfs[x] = df_country_world_happiness_merged.loc[df_country_world_happiness_merged['Region']==x]
```

```python
regions_dfs['Caribbean'].dtypes
```

```python
#reindex and print the happiest country in each region
for x in regions_dfs:
    regions_dfs[x].reset_index(inplace=True)
    del regions_dfs[x]['index']
    print(regions_dfs[x].loc[0,'country'])
```

```python
#Generating the Data Load SQL file for Postgres
for reg in regions_dfs:
    file = open('regiontables/'+reg+".sql","w")
    for index, row in regions_dfs[reg].iterrows():
        values=""
        for x in cols:
            if x != 'Net Official Development Assist. received (% of GNI)':
                values=values+"'"+str(row[x]+"', "
            else:
                values=values+"'"+str(row[x])+"' "
        file.write("insert into " + reg + " (" + collumns+")"+ "values (" + values + ");"+"\n")
    file.close()
```
